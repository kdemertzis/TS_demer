```
import pycaret
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pycaret.classification import *
```

```
# define the location of the dataset
url = 'https://raw.githubusercontent.com/kdemertzis/TS_demer/main/Data/Fed.csv'
# load the dataset
df = pd.read_csv(url, header=None)
# summarize the shape of the dataset
print(df.shape)
```

```
# set column names as the column number
n_cols = df.shape[1]
df.columns = [str(i) for i in range(n_cols)]
```

```
# summarize the first few rows of data
print(df.head())
```

```
# setup the dataset
grid = setup(data=df, target=df.columns[-1], html=False, silent=True, verbose=False)
```

```
best = compare_models(cross_validation=False, n_select = 1) 
%load_ext tensorboard
```

```
# report the best model
print(best)
```

```
evaluate_model(best)
```

```
# save model to drive
save_model(best, 'my_best')
```

```
# save model to drive
save_model(best, 'best_model')
```

```
# load the model
from pycaret.classification import load_model
loaded_model = load_model('best_model')
```

```
# generate predictions / inference
from pycaret.classification import predict_model
pred = predict_model(loaded_model, data=df) # new data
```

```
# convert best model to onnx
from skl2onnx import to_onnx
X_sample = get_config('X_train')[:1]
model_onnx = to_onnx(best, X_sample.to_numpy())
```

```
# save the model to drive
with open("best_model.onnx", "wb") as f:
    f.write(model_onnx.SerializeToString())
```

```
# generate inference on onnx
from onnxruntime import InferenceSession
sess = InferenceSession(model_onnx.SerializeToString())
X_test = get_config('X_test').to_numpy()
predictions_onnx = sess.run(None, {'X': X_test})[0]

# print predictions_onnx
print(predictions_onnx)
```
